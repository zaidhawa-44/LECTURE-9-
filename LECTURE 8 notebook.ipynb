```python
import pandas as pd
```

**SCORE PROCESSOR**

```python
def get_scores_df(report_card_dictionary, division_name):
    
    #Convert the dictionary into DataFrame
    report_card_df= pd.DataFrame(report_card_dictionary)
    #Add the total Score column
    report_card_df["total score"]= report_card_df["Math_scores"]+report_card_df["chem_scores"]+report_card_df["IT_scores"]+report_card_df["bio_scores"]
    #Add the average score column
    report_card_df["average score"]= report_card_df["total score"]/4
    #Add the pass fail column
    report_card_df.loc[report_card_df["total score"]>=140,"status"]= "pass"
    report_card_df.loc[report_card_df["total score"]<140,"status"]= "fail"
    #Add the character Grade column
    report_card_df.loc[report_card_df["average score"]>=90,"Character Grade"]="A"
    report_card_df.loc[(report_card_df["average score"]>=70)&(report_card_df["average score"]<90), "Character  Grade"] = "B"
    report_card_df.loc[(report_card_df["average score"]>=50)&(report_card_df["average score"]<70), "Character  Grade"]= "C"
    report_card_df.loc[(report_card_df["average score"]>=35)&(report_card_df["average score"]<50), "Character  Grade"] = "D"
    report_card_df.loc[report_card_df["average score"]<=35, "Character Grade"] = "F"
    #Add the division column
    report_card_df["division"]= division_name
    
    #Return the report card data frame
    return report_card_df
```

**Get email function**

```python
def get_emails_df(emails_dictionary, division_name):
    
    # Converting our dictionary into DataFrame
    emails_df = pd.DataFrame(emails_dictionary)
    
    # Add a division column to our email dataframe
    emails_df["division"] = division_name
    
    # Converting our emails to lower case
    emails_df["student email"] = emails_df["student email"].str.lower()
    
    # Return the email dataframe
    return emails_df
```

```python
div_a_emails_df = get_emails_df(div_a_email_dictionary, "A")
div_a_emails_df
```

```python
def get_scores_emails_df(scores_dictionary, emails_dictionary, division_name):
    #Step1 Make the score dictionary into dataframe scorecard
    score_card_df = get_scores_df(scores_dictionary,division_name)
    
    #Step 2 Make email dictionary into email dataframe
    emails_df = get_emails_df(emails_dictionary,division_name)
    
    # Step 3 Merge the data frame
    score_card_email_df = pd.merge(score_card_df, emails_df, how= "left", on = ["student","division"])
    
    #Step4 Return the merged frame
    return score_card_email_df
```

```python
get_scores_emails_df(div_a_score_dictionary,div_a_email_dictionary, "A")
```

```python
def get_all_division_df(div_1_scores_dictionary, div_1_emails_dictionary, div_1_name,
                        div_2_scores_dictionary, div_2_emails_dictionary, div_2_name,
                        div_3_scores_dictionary, div_3_emails_dictionary, div_3_name ):
    
    div_1_df = get_scores_emails_df(div_1_scores_dictionary, div_1_emails_dictionary, div_1_name)
    div_2_df = get_scores_emails_df(div_2_scores_dictionary, div_2_emails_dictionary, div_2_name)
    div_3_df = get_scores_emails_df(div_3_scores_dictionary, div_3_emails_dictionary, div_3_name)
    
    all_div_df = pd.concat([div_1_df, div_2_df, div_3_df], ignore_index=True)
    
    return all_div_df
```

```python
def save_all_division_df(div_1_scores_dictionary, div_1_emails_dictionary, div_1_name,
                        div_2_scores_dictionary, div_2_emails_dictionary, div_2_name,
                        div_3_scores_dictionary, div_3_emails_dictionary, div_3_name ):
    
    div_1_df = get_scores_emails_df(div_1_scores_dictionary, div_1_emails_dictionary, div_1_name)
    div_2_df = get_scores_emails_df(div_2_scores_dictionary, div_2_emails_dictionary, div_2_name)
    div_3_df = get_scores_emails_df(div_3_scores_dictionary, div_3_emails_dictionary, div_3_name)
    
    all_div_df = pd.concat([div_1_df, div_2_df, div_3_df], ignore_index=True)
    
    all_div_df.to_csv("All Division Scorecard.csv")
```
